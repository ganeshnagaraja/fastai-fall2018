```python
from fastai import *
from fastai.vision import *
```

```python
# batch size (reduce this if you run out of memory!)
bs = 64
```

```python
path = untar_data(URLs.PETS)
```

```python
print(path)
```

```python
path.ls()
```

```python
path_anno = path/'annotations'
```

```python
path_anno
```

```python
path_img = path/'images'
```

```python
doc(get_image_files)
```

```python
fnames = get_image_files(path_img, True)
```

```python
fnames[:5]
```

```python
len(fnames)
```

```python
np.random.seed(2)
pat = r'/([^/]+)_\d+.jpg$'
```

```python
doc(ImageDataBunch)
```

```python
doc(ImageDataBunch.from_name_re)
```

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size=224, bs=bs)
```

```python
data.normalize(imagenet_stats)
```

```python
data.show_batch(rows=4, figsize=(7,6))
```

## Time to train our model

```python
learn = ConvLearner(data, models.resnet34, metrics=error_rate, path="")
```

```python
print(learn.model_dir)
print(URLs.LOCAL_PATH)
```

```python
learn.save('foo')
```

```python
learn.load('foo')
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9, figsize=(15, 11))
```

```python
doc(interp.plot_top_losses)
```

```python
interp.plot_confusion_matrix(figsize=(12, 12), dpi=60)
```

```python
interp.most_confused(min_val=2)
```

```python
doc(learn.unfreeze)
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(1)
```

```python
learn.load('stage-1')
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-6, 1e-4))
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9, figsize=(15, 11))
```

```python
interp.most_confused(min_val=2)
```

### Training with resnet50

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size=320, bs=bs//2)
data.normalize(imagenet_stats)
```

```python
bs//2
```

```python
bs/2
```

```python
learn = ConvLearner(data, models.resnet50, metrics=error_rate)
```

```python
learn.fit_one_cycle(8, max_lr=slice(1e-3))
```

```python
learn.save('stage-1-50')
```
